```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = 'C:\\Users\\Admin\\Desktop\\Vandana\\Projects\\MatplotlibLearnings\\MatplotlibLearnings\\iplauction2023.csv'
data = np.genfromtxt(filepath, delimiter=',', dtype=None, encoding=None, names=True)
#print("DATA: ", data)
column_name = 'nationality'
def count_roles_by_nationality(data, nation):
    # Filter data by nationality
    filtered_data = data[data[column_name] == nation]
   # print("Filter data: ", filtered_data)
  
    # Count bowlers and batsmen
    bowlers_count = np.sum(filtered_data['player_style'] == 'Bowler')
    batsmen_count = np.sum(filtered_data['player_style'] == 'Batter')
    all_rounder = np.sum(filtered_data['player_style'] == 'Allrounder')
    return nation,bowlers_count, batsmen_count, all_rounder
```

```python

nationalityData = [ 'Afghanistan', 'Australia', 'Bangladesh', 'England', 'India', 'Ireland', 'Namibia', 'Netherlands', 'New Zealand', 'Singapore', 'South Africa',
  'Sri Lanka', 'West Indies', 'Zimbabwe']
bowlerCount = np.empty((0,))
batsmenCount = np.empty((0,))
allRounderCount =np.empty((0,))
for nation in nationalityData:
    nation, bowlers_count, batsmen_count, all_rounder = count_roles_by_nationality(data,nation)
    bowlerCount = np.append(bowlerCount,bowlers_count)
    batsmenCount=np.append(batsmenCount,batsmen_count)
    allRounderCount=np.append(allRounderCount,all_rounder)
    
print("Nation@: ", nationalityData)
print("bowlers_count@: ", bowlerCount)
print("batsmen_count@: ", batsmenCount)
print("all_rounder@: ", allRounderCount)
```

```python
plt.bar(nationalityData,allRounderCount, color ='r')
plt.bar(nationalityData,bowlerCount,bottom=allRounderCount, color ='g')
plt.bar(nationalityData,batsmenCount,bottom =bowlerCount+allRounderCount,color ='y')
plt.xticks(rotation = 30,ha = 'right' )
#plt.tight_layout()
```
