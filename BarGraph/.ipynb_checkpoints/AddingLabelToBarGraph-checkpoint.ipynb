```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
beverages = np.array(['Water','Tea','Coffee','Orange Juice','Lemon Juice']) # x-axis
people = np.array([45,35,25,10,40])# y-axis
```

```python
# Adding Text label in bar with horizontal & vertical alignment
colors = ['orange','red','blue','green','grey']
plt.bar(beverages,people,color=colors,width =0.5)
for i in range(len(beverages)):
    textValue = '>30' if people[i]>30 else '<30'
    plt.text(beverages[i],people[i],textValue,color = 'brown',ha = 'center',va = 'bottom',backgroundcolor='#FAECDA')
plt.tight_layout()
```

```python
# Adding Annotate label in bar with horizontal & vertical alignment
colors = ['orange','red','blue','green','grey']
plt.bar(beverages,people,color=colors,width =0.5)
for i in range(len(beverages)):
    textValue = '>30' if people[i]>30 else '<30'
    plt.annotate(textValue,(beverages[i],people[i]),color = 'cyan',ha = 'center',va = 'bottom')
plt.tight_layout()
```
