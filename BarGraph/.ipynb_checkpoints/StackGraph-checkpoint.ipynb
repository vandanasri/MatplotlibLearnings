```python
import matplotlib.pyplot as plt
import numpy as np

studentRollNo = ['Rn01','Rn02','Rn03','Rn04','Rn05','Rn06']
practicalMarks = [78,60,83,49,96,35]
testMarks = [69,56,89,50,85,58]

# created stack graph using array
plt.bar(studentRollNo,practicalMarks, color='r')
plt.bar(studentRollNo,testMarks, bottom=practicalMarks,color='g')
plt.grid(axis='y')
plt.xlabel('Roll Number')
plt.ylabel('Students Marks')
plt.title('Total Marks Scored')
```

```python

english = np.array([78,60,83,49,96,35])
maths = np.array([69,56,89,50,85,58])
science = np.array([78,90,64,56,77,62])
socialStudy = np.array([61,76,83,66,69,48])
# created vertical stack graph using ndarray, it is must to use ndarray for bottom addition, as it does element addition.
plt.bar(studentRollNo,english, color ='orange', label='English')
plt.bar(studentRollNo,maths,bottom =english, color ='grey', label='Maths')
plt.bar(studentRollNo,science, bottom = english + maths, color='b', label='Science')
plt.bar(studentRollNo,socialStudy, bottom = english + maths+science, color='brown', label='Social Study')
plt.xlabel('Roll Number')
plt.ylabel('Students Marks')
plt.title('Total Marks Scored')
plt.legend(loc=(0.8,1.1))
plt.tight_layout()
# adding annotation
for i in range(len(studentRollNo)):
    total = english[i]+maths[i]+science[i]+socialStudy[i]
    plt.annotate(str(total),(studentRollNo[i],total+2), ha='center')
    plt.annotate(str(english[i]),(studentRollNo[i],english[i]/2), ha='center')
    plt.annotate(str(maths[i]),(studentRollNo[i],english[i]+maths[i]/2), ha='center')
    plt.annotate(str(science[i]),(studentRollNo[i],english[i]+maths[i]+science[i]/2), ha='center')
    plt.annotate(str(socialStudy[i]),(studentRollNo[i],english[i]+maths[i]+science[i]+socialStudy[i]/2), ha='center')
```

```python
# created Horizontal stack graph using ndarray, it is must to use ndarray for bottom addition, as it does element addition.
plt.barh(studentRollNo,english, color ='orange',label='English')
plt.barh(studentRollNo,maths,left =english, color ='grey',label='Maths')
plt.barh(studentRollNo,science, left = english + maths, color='b', label='Science')
plt.barh(studentRollNo,socialStudy, left = english + maths+science, color='brown', label='Social Study')
plt.xlabel('Roll Number')
plt.ylabel('Students Marks')
plt.title('Total Marks Scored')
plt.legend(ncol=4, loc = (0,1.1))
# adding annotation
for i in range(len(studentRollNo)):
    total = english[i]+maths[i]+science[i]+socialStudy[i]
    plt.annotate(str(total),(total+2,studentRollNo[i]),va='center')
    plt.annotate(str(english[i]),(english[i]/2,studentRollNo[i]),ha='center',  va='center')
    plt.annotate(str(maths[i]),(english[i]+maths[i]/2,studentRollNo[i]), ha='center', va='center')
    plt.annotate(str(science[i]),(english[i]+maths[i]+science[i]/2,studentRollNo[i]),ha='center', va='center')
    plt.annotate(str(socialStudy[i]),(english[i]+maths[i]+science[i]+socialStudy[i]/2,studentRollNo[i]),ha='center', va='center')
```
