```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw the horizontal bar graph for the given data set between the number of people and their preferred beverage.
beverages = np.array(['Water','Tea','Coffee','Orange Juice','Lemon Juice']) # x-axis
people = np.array([45,35,25,10,40])# y-axis
plt.bar(beverages,people,color='orange')
plt.title('Preferred Beverages',color='r',size=15)
plt.xlabel('Beverages',color='g', size=12)
plt.ylabel('Number of People', color='g',size=12)
```

```python
# different colors for different bars, with customised width as 0.5
colors = ['orange','red','blue','green','grey']
plt.bar(beverages,people,color=colors,width =0.5)
plt.title('Preferred Beverages',color='r',size=15)
plt.xlabel('Beverages',color='g', size=12)
plt.ylabel('Number of People', color='g',size=12)
plt.xticks(beverages, rotation = 35)
```

```python
# customize different width for different bars
colors = ['orange','red','blue','green','grey']
widthValues = [0.5,0.8,0.3,0.1,1]
plt.bar(beverages,people,color=colors,width = widthValues )
```

```python
# different colors reused
colors = ['orange','red']
plt.bar(beverages,people,color=colors )
```

```python
# customize different bottom values i.e. start values for different bars, we can pass single value also
colors = ['orange','red','blue','green','grey']
bottomValues = [0,40,50,20,10]
plt.bar(beverages,people,color=colors,width =0.5, bottom = bottomValues)
```

```python
# customize alignment if take width negative and align = edge then bar moves towards Left else right 
colors = ['orange','red','blue','green','grey']
widthValues = [-0.5,0.8,0.3,-0.1,1]
plt.bar(beverages,people,color=colors,width = widthValues, align = 'edge' )
```

```python
# adjust everything in the graph by using tight_layout
colors = ['orange','red','blue','green','grey']
widthValues = [-0.5,0.8,0.3,-0.1,1]
plt.bar(beverages,people,color=colors,width = widthValues, align = 'edge' )
plt.tight_layout()
```
