```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
studentRollNo = ['Rn01','Rn02','Rn03','Rn04','Rn05','Rn06']
english = np.array([78,60,83,49,96,35])
maths = np.array([69,56,89,50,85,58])
science = np.array([78,90,64,56,77,62])
socialStudy = np.array([61,76,83,66,69,48])
```

```python
# draw cluster or grouped bar chart
xpos = np.arange(len(studentRollNo))
w =0.2
plt.bar(xpos,english,color='r',width=w)
plt.bar(xpos+w,maths,color='g',width=w)
plt.bar(xpos+2*w,science, color ='b',width=w)
plt.bar(xpos+3*w,socialStudy, color ='y',width=w)
plt.xticks(xpos+3*w/2, studentRollNo)
plt.tight_layout()
```
