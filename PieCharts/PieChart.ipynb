```python
import matplotlib.pyplot as plt
import numpy as np

studentRollNo = ['Rn01','Rn02','Rn03','Rn04','Rn05','Rn06']
totalMarks = [78,60,83,49,96,35]

# creating pie chart with auto percentage
plt.pie(totalMarks, autopct='%.1f%%',labels = studentRollNo)
```

```python
# creating pie chart with explode and shadow = True
explodeValue =[0.0,0.0,0.2,0.0,0.0,1.0]
plt.pie(totalMarks, autopct='%.1f%%',labels = studentRollNo, explode=explodeValue, shadow=True)
```

```python
# change color of pie chart, by default it is anticlockwise, so make it clockwise by using counterclock = False
colorValues=['#769eb9','#447130','#ed8a80','#fee020','#adbc40','r']
plt.pie(totalMarks,labels=studentRollNo,autopct='%.1f%%', colors=colorValues,shadow=True, counterclock=False)
```

```python
# creating pie chart with startangle=60 and adding legend
explodeValue =[0.0,0.0,0.2,0.0,0.0,0.0]
plt.pie(totalMarks, autopct='%.1f%%',labels = studentRollNo, explode=explodeValue, startangle=60)
plt.legend(title='Students Marks', loc =(1,1))
plt.title('Total Marks Scored')
```

```python
#change the edge property of chart
colorValues=['#769eb9','#447130','#ed8a80','#fee020','#adbc40','r']
plt.pie(totalMarks,labels=studentRollNo,autopct='%.1f%%', colors=colorValues, wedgeprops={'edgecolor':'k','linestyle':':','linewidth':2})
```
