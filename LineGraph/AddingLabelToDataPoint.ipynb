```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,20,2)
a
```

```python
# Add Text to dataPoints
plt.plot(a,a*10, 'o-b',label='a*10')
for i in range(a.size):
    plt.text(a[i]+0.3,a[i]*10-2,f'{(a[i],a[i]*10)}',size = 6, color='g')
```

```python
# Annotate to dataPoints
plt.plot(a,a*10, 'o-b',label='a*10')
for i in range(a.size):
    plt.annotate(f'{(a[i],a[i]*10)}',(a[i]+0.3,a[i]*10-2),size = 6, color='r')
```
