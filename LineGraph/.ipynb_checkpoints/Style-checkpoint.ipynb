```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,20)
a
```

```python
# plot style='ggplot'
plt.title('Single Graph Multiple Plots')
plt.style.use('ggplot')
plt.plot(a,a*10,'*:')
```

```python
# plot style='seaborn'
plt.title('Single Graph Multiple Plots')
plt.style.use('fivethirtyeight')
plt.plot(a,a*10,'*:')
```
