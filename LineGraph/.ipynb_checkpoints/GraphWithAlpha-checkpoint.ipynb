```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create 2-D array
a = np.arange(1,11)
a
```

```python
plt.title("Line with Transparancy")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding mlc and alpha = 0.5 transparent
plt.plot(a,a**2,'o-r',alpha=0.5)
```

```python
plt.title("Line Graph with LineWidth Marker")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding mlc and alpha = 0.5 transparent
plt.plot(a,a**2,'o-r',alpha=0.1,lw = 5)
```
