```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,20)
a
```

```python
# show grid when boolean argument is none, toggles the visibility of lines
plt.plot(a,a*10, marker='o')
plt.grid() #visible
```

```python
# Invisible grid when boolean argument is none, toggles the visibility of lines
plt.plot(a,a*10, marker='o')
plt.grid() 
plt.grid() #invisible
```

```python
# show grid when boolean argument is none, but passing keyword argument
plt.plot(a,a*10, marker='o')
plt.grid()
plt.grid(color='#DB710E')  #visible, beacuse keyword argument has passed
```

```python
# show grid when boolean argument is True
plt.plot(a,a*10, marker='o')
plt.grid(True)
```

```python
# show grid which argument as major
plt.plot(a,a*10, marker='o')
plt.grid(True, which='major',color='red')
```

```python
# show grid which argument as minor
plt.plot(a,a*10, marker='o')
plt.minorticks_on()
plt.grid(True, which='minor',color='red')
```

```python
# show grid which argument as minor, with minorticks_off
plt.plot(a,a*10, marker='o')
plt.minorticks_off()
plt.grid(True, which='minor',color='red')
```

```python
# show grid which argument as both
plt.plot(a,a*10, marker='o')
plt.minorticks_on()
plt.grid(True, which='both',color='#C80EDB')
```

```python
# show grid with major and minor lines different colors
plt.plot(a,a*10, marker='o',color='#DB0E62')
plt.grid(color='#620EDB',lw=2)
plt.minorticks_on()
plt.grid(True, which='minor',color='#C80EDB')
```

```python
#show only y-axis in grid 
plt.plot(a,a*10, marker='o',color='#DB0E62')
plt.grid(color='#620EDB',lw=2, axis='y')
plt.minorticks_on()
plt.grid(True, which='minor',color='#C80EDB', axis='y')
```

```python
#show only x-axis in grid 
plt.plot(a,a*10, marker='o',color='#DB0E62')
plt.grid(color='#620EDB',lw=2, axis='x')
plt.minorticks_on()
plt.grid(True, which='minor',color='#C80EDB', axis='x')
```

```python
#show grid linestyle 
plt.plot(a,a*10, marker='o',color='#DB0E62')
plt.grid(color='#620EDB',lw=2, linestyle=':')
```
