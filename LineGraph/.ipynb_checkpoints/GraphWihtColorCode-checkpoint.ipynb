```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create 2-D array
a = np.arange(1,11)
a
```

```python
b = a**2
b
```

```python
plt.title("Line Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='o', linestyle='--', color='magenta'
plt.plot(a,b,marker='o', linestyle='--', color='m')
```

```python
plt.title("Line Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='o', color='green'
plt.plot(a,b,marker='o', color='green')
```

```python
plt.title("Line Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='o', linestyle='--', color='#FF3383'
plt.plot(a,b,marker='o', linestyle='--', color='#FF3383')
```

```python
#plot graph with default colors, if there are more than one lines
plt.plot(a,a*2)
plt.plot(a,a*8)
plt.plot(a,a*16)
plt.plot(a,a*32)
plt.plot(a,a*64)
plt.title("Line with Default Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```

```python
#formatted style mlc
plt.title("Line Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='o', linestyle='--', color='#FF3383'
plt.plot(a,b,'o--g')
```

```python
#formatted style mlc with marker face color (mfc)
plt.title("Line Color Property")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='o', linestyle='--', color='#FF3383'
plt.plot(a,b,'o--g',mfc='r')
```
