```python
import matplotlib.pyplot as plt
import numpy as np

# create 2-D array
a = np.arange(1,11)
a
```

```python
#Saved in Default path
plt.figure(num='Fig1',figsize=(3,2),dpi=200.6,facecolor='#33FFDD')
plt.plot(a,a**10)
plt.savefig('LineGraph_DefaultPath.png')
```
