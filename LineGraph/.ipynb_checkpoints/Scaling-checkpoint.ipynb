```python
import matplotlib.pyplot as plt
import numpy as np

# create 2-D array
a = np.arange(1,11)
a
```

```python
# set the x-axis scale and y-scale
plt.title('Single Graph Multiple Plots')
plt.plot(a,a*1000,'*:')
plt.grid()
plt.xscale('linear')
plt.yscale('log')
```

```python
# set the x-axis scale and y-scale with customized base value
plt.title('Single Graph Multiple Plots')
plt.plot(a,a*1000,'*:')
plt.grid()
plt.xscale('log', base=3)
plt.yscale('log', base =2)
```
