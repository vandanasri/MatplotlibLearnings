```python
import matplotlib.pyplot as plt
import numpy as np

# create 2-D array
a = np.arange(1,11)
a
```

```python
b = a**2
b
```

```python
plt.title('Line Graph with Dashed Line Style')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
#adding marker property as marker ='o' & linestyle='--'
plt.plot(a,b,marker='o', linestyle ='--')
```

```python
plt.title('Line Graph with Solid Line Style')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
#adding marker property as marker ='o' & linestyle='-'
plt.plot(a,b,marker='o', linestyle ='-')
```

```python
plt.title('Line Graph with Dash-Dot Line Style')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
#adding marker property as marker ='o' & linestyle='-.'
plt.plot(a,b,marker='o', linestyle ='-.')
```

```python
plt.title('Line Graph with Dotted Line Style')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
#adding marker property as marker ='o' & linestyle=':'
plt.plot(a,b,marker='o', linestyle =':')
```
