```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,20)
a
```

```python
# plot in multiple plots
plt.title('Single Graph Multiple Plots')
plt.plot(a,a*10,'*:')
plt.plot(a,a**2,'.--g')
plt.plot(a,'s-r')
```

```python
# plot in by giving single plot
plt.title('Single Graph Single Plots')
plt.plot(a,a*10,'*:',a,a**2,'.--g',a,'s-r')
```

```python
# plot in by giving single plot with alpha =0.5
plt.title('Single Graph Single Plots')
plt.plot(a,a*10,'*:',a,a**2,'.--g',a,'s-r',alpha = 0.5)
```

```python
# plot in by giving single plot with alpha =0.5 and linewidth =5
plt.title('Single Graph Single Plots')
plt.plot(a,a*10,'*:',a,a**2,'.--g',a,'s-r',alpha = 0.5, linewidth = 5)
```
