```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create graph from 1-D array
a= np.array([10,20,30,40,50])
#here a is y-axis, x-axis value is genegrated automatically
plt.plot(a,marker='.')
plt.title("Line Graph from 1-D array")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```

```python
# create 2-D array
a = np.arange(1,11)
a
```

```python
b = a**2
b
```

```python
plt.plot(a,b)
plt.title("Line Graph")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```

```python
# to get default sizee of Figure
plt.rcParams['figure.figsize']
```

```python
#customize size of Figure and change facecolor to green
plt.figure(num='Fig1',figsize=(2,2),facecolor='green')
plt.plot(a,b)
```

```python
#customize Figure with frameon=False by default it is true
plt.figure(num='Fig1',figsize=(3,2),facecolor='green',frameon=False)
plt.plot(a,b)
```

```python
#customize Figure with frameon=True and dpi=200.6
plt.figure(num='Fig1',figsize=(3,2),dpi=200.6,facecolor='blue',frameon=True)
plt.plot(a,b)
```
