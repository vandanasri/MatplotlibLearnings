```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
a = np.arange(1,20)
a
```

```python
plt.title('Plotting Customized X & Y Graph',loc='center', pad =10)
xdict={'color':'White','fontstyle':'italic','size':'12','alpha':1, 'backgroundcolor':'#0EDBCF', 'weight':10, 'rotation':1}
ydict={'fontstyle':'italic','size':'12','alpha':0.5, 'backgroundcolor':'#9ADB0E', 'weight':10, 'rotation':90}
plt.xlabel('Here is X label',xdict, labelpad=10)
plt.ylabel('Here is Y label',ydict,color='black', labelpad=10)
plt.plot(a,a*10, marker='o')
```
