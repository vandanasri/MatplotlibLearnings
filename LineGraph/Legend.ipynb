```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(1,20)
a
```

```python
#legend
plt.plot(a,a*10, 'o-b',label='a*10')
plt.plot(a,a*20, '.-g',label='a*20')
plt.plot(a,a*30,'*-y',label='a*30')
plt.legend()
```

```python
# legend([labels]) ==> must know the order
plt.plot(a,a*10, 'o-b',label='a*10')
plt.plot(a,a*20, '.-g',label='a*20')
plt.plot(a,a*30,'*-y',label='a*30')
plt.legend(['a*30','a*10','a*20'])
# below the legend doesn't match with the line graph, order mismatch so we have to know the order of the graph plotted
```

```python
# legend([handles],[labels])
line1, = plt.plot(a,a*10, 'o-b',label='a*10')
line2, = plt.plot(a,a*20, '.-g',label='a*20')
line3, = plt.plot(a,a*30,'*-y',label='a*30')
plt.legend([line1,line2,line3],['a*10','a*20','a*30'])
```

```python
# legend([handles],[labels]) another way
line1,line2,line3 = plt.plot(a,a*10, 'o-b',a,a*20, '.-g',a,a*30,'*-y')
plt.legend([line1,line2,line3],['a*10','a*20','a*30'])
```

```python
# legend(handles,[labels]) another way, with location
lines = plt.plot(a,a*10, 'o-b',a,a*20, '.-g',a,a*30,'*-y')
plt.legend(lines,['a*10','a*20','a*30'],loc=9)
```

```python
#  with location & number of columns
lines = plt.plot(a,a*10, 'o-b',a,a*20, '.-g',a,a*30,'*-y')
plt.legend(lines,['a*10','a*20','a*30'],loc=9, ncol = 3)
```

```python
#  with title
lines = plt.plot(a,a*10, 'o-b',a,a*20, '.-g',a,a*30,'*-y')
plt.legend(lines,['a*10','a*20','a*30'],title='Three Line Plots',loc=9, ncol = 3)
```

```python
#  legend outside the graph
lines = plt.plot(a,a*10, 'o-b',a,a*20, '.-g',a,a*30,'*-y')
plt.legend(lines,['a*10','a*20','a*30'],title='Three Line Plots',loc=(0,1.1), ncol = 3)
```
