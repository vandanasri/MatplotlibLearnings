```python
import matplotlib.pyplot as plt
import numpy as np

# create 2-D array
a = np.arange(1,11)
a
```

```python
b = a**2
b
```

```python
plt.title("Line Graph Point Marker")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker ='.'
plt.plot(a,b,marker='.')
```

```python
plt.title("Line Graph Pixel Marker")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#adding marker property as marker =','
plt.plot(a,b,marker=',')
```
