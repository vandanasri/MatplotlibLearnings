```python
import matplotlib.pyplot as plt
import numpy as np

a=[1,2,3,4,5]
b=[10,20,30,40,50]
l=['A','B','C','D','E']
```

```python
# create figure
fig = plt.figure(1,(10,10))
```

```python
fig = plt.figure(1,(7,7))
plt.suptitle('Four different graphs', color='r', size=15)
# Axis One
axis1= plt.subplot(221) # can write this also means (2,2,1)
axis1.plot(a,b,marker='o')
axis1.set(title ='Line Graph', xlabel = 'Axis-1 X', ylabel = 'Axis-1 y')
plt.tight_layout()
# Axis Two
axis2= plt.subplot(2,2,2)
axis2.bar(a,b)
axis2.set_facecolor('yellow')
axis2.set_title('Bar Graph',color='b')
axis2.set_xlabel('Axis-2 X', color='b')
axis2.set_ylabel('Axis-2 Y', color='b')
plt.tight_layout()

# Axis Three
axis3= plt.subplot(2,2,3)
explodeValue =[0.0,0.0,0.0,0.2,0.0]
axis3.pie(a,labels=l,autopct='%.1f%%',explode = explodeValue )
axis3.set(title ='Pie Chart')
plt.tight_layout()

# Axis Four
axis4= plt.subplot(2,2,4)
v1 = np.random.randint(1,10, size=10)
v2 = np.random.randint(21,30, size=10)
size =  np.random.randint(50,60, size=10)
colors =np.random.randint(91,100,10)
axis4.scatter(v1,v2,s=size, c=colors,cmap = 'prism', alpha=0.6)
axis4.set(title ='Scatter Graph', xlabel = 'Axis-4 X', ylabel = 'Axis-4 y')
```

```python
#return one axes and one figure
fx,ax=plt.subplots()
fx
ax
```

```python
# return 2 axes and one figure
fx,ax=plt.subplots(2,2)
fig
ax
```

```python
# plot graph with subplots

fx,ax=plt.subplots(2,2)
ax[0][0].plot(a,b,marker='o')
ax[0][1].bar(a,b)
ax[1][0].pie(a,labels=l,autopct='%.1f%%',explode = explodeValue )
ax[1][1].scatter(v1,v2,s=size, c=colors,cmap = 'prism', alpha=0.6)
plt.suptitle('Graph with Subplots', color='b', size=15)
```

```python
# Unboxing 
fx,((axOne,axTwo),(axThree,axFour))=plt.subplots(2,2)
fx.set_size_inches(8,8) # set the size of figure
axOne.plot(a,b,marker='o')
axTwo.bar(a,b)
axThree.pie(a,labels=l,autopct='%.1f%%',explode = explodeValue )
axFour.scatter(v1,v2,s=size, c=colors,cmap = 'prism', alpha=0.6)
plt.suptitle('Graph with Unboxing Subplots', color='b', size=15)
```
