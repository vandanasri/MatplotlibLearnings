```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
 # Uniform distribution of 1000 values in the interval [0,1)
value = np.random.rand(1000)
#creating histogram, with edge color black
 plt.hist(value, ec='k')
```

```python
#check number of patches and value size in histogram
valuesInEachBar,bins,patches=plt.hist(value, ec='k', color='#709168')
print(f"No of values in each bar ${valuesInEachBar}")
print(f'The edge of the bins ${bins}')
print(f'The Patches-container ${patches}')
print(f'The Value size ${valuesInEachBar.size}')
```

```python
# as the value size is 10, so, take 10 colors
print(f'The Patches-container ----->${patches}')
colors=  ['red', 'green', 'blue', 'yellow', 'purple', 'orange', 'cyan', 'magenta', 'brown', 'pink']
valuesInEachBar,bins,patches=plt.hist(value, ec='k',bins=len(colors))
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)
```
