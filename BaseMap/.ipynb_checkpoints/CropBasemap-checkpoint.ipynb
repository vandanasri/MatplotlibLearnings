```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
lowerleftCornerLatiude = 7.8900
lowerleftCornerLongitude = 64.6128
upperrightCornerLatitude = 39.3174
upperrightCornerLongitude = 97.6412
latPlaceOnMap = [27.1571,17.3616,28.6562, 13.6288, 19.7645]
lonPlaceOnMap = [78.0421, 78.4747, 77.2410, 79.4192, 74.4762]
placesName = ['Taj Mahal', 'Char Minar', 'Red Fort', 'Tirumala', 'Shirdi']
```

```python
# crop particular country through latitude and longitude.
fig = plt.figure(figsize=(10,20))
m=Basemap(projection = 'mill', resolution ='c', llcrnrlat=lowerleftCornerLatiude, llcrnrlon = lowerleftCornerLongitude ,urcrnrlat= upperrightCornerLatitude, urcrnrlon= upperrightCornerLongitude)
m.drawcoastlines(color = 'b')
m.drawcountries(color = 'brown')
m.drawmapboundary(color = 'orange')
# plotting different points in map with their latitude and longitude
x,y = m(lonPlaceOnMap,latPlaceOnMap)
plt.scatter(x, y, s=200, c=[0,20,30,40,60], cmap='prism')
for index,name in enumerate(placesName):
    plt.annotate(name,(x[index]+ 100000, y[index]))
```
