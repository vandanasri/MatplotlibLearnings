```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib widget
fig = plt.figure()
ax = plt.axes(projection = '3d')
x = np.random.randint(1,61, size=60)
y = np.random.randint(99,800, size=60)
z = np.random.randint(800,1000, size=60)
colors =np.random.randint(1,100,60)
ax.scatter3D(x,y,z, c=colors, cmap='prism',s=50)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')
ax.set_title('3-D Scatter Graph')
#plt.close()
```
