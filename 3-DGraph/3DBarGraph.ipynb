```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib widget
fig = plt.figure()
ax = plt.axes(projection = '3d')
x =[1,2,3,4,5,6]
y = [6,8,3,9,2,7]
z = np.ones(6)
dx = np.ones(6)
dy = np.ones(6)
dz = [1,2,3,4,5,6]
ax.bar3d(x,y,z,dx,dy,dz, color=['r','b','g','orange','grey','y'])
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')
ax.set_title('3-D Bar Graph')
#plt.close()
```
