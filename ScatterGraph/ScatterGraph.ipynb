```python
import matplotlib.pyplot as plt
import numpy as np

x = [10,20,30,40,50,60]
y = [100,200,300,400,500,600]
```

```python
#plot scatter graph with differemt size and shape
plt.scatter(x,y,color=['r','g','b','grey','yellow','orange'], s=[120,100,300,500,800,50])
```

```python
#plot scatter graph with differemt size and colormap
v1 = np.random.randint(1,61, size=60)
v2 = np.random.randint(99,800, size=60)
size =  np.random.randint(100,300, size=60)
colors =np.random.randint(1,100,60)
plt.scatter(v1,v2,s=size, c=colors,cmap = 'prism', alpha=0.6)
plt.colorbar() # to show colorbar
```
